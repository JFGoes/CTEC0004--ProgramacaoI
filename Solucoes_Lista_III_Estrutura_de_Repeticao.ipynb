Faça um programa que calcule e escreva o número de anos necessários para que a população do país A ultrapasse ou iguale a população do país B, mantidas as taxas de crescimento.

```
# Solução: função linear: f(x) = po + i * po, 
# sendo po a popuçação inicial e i a taxa de crescimento
pop_A = 80000
taxa_A = 0.03
pop_B = 200000
taxa_B = 0.015 
ano = 0
while pop_A < pop_B:
   pop_A += pop_A * taxa_A
   pop_B += pop_B * taxa_B
   ano+=1
else:
  print(f"Serão necessários {ano} anos para a população do país A ser maior ou igual a B")
```

##42 --	Faça um programa que gere aleatoriamente, 1000 números positivos e conte quantos deles estão nos seguintes intervalos: [0-25], [26-50], [51-75] e [76-100]. A entrada de dados deverá terminar quando for o contador atingir um número maior que 1000.

```
import random
import numpy as np
```

```
count1 = 0
count2 = 0
count3 = 0
count4 = 0
cont = 0
while cont < 1000:
  num = random.randint(0,100)
  #print(num)
  if num>=0 and num <= 25:
    count1 += 1
  elif num>=26 and num <= 50:
    count2 += 1
  elif num>=51 and num <= 75:
    count3 += 1
  elif num>=76 and num <= 100:
    count4 += 1
  else:
    print("valor fora do intervalo")
  cont+=1

print("[0-25]: {}".format(count1))
print ("[26-50]: {}".format(count2))
print ("[51-75]: {}".format(count3))
print ("[76-100]: {}".format(count4))
```
